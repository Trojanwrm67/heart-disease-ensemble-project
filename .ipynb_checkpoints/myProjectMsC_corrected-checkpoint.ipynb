```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.base import clone

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier, BaggingClassifier

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, confusion_matrix,
    ConfusionMatrixDisplay, classification_report
)

RANDOM_STATE = 42
print("All imports done")
```

```python

DATA_PATH = "heart_disease_uci.csv"
raw_df = pd.read_csv(DATA_PATH)


# Replace any '?' with NaN then convert num to binary
# 0 = no disease, 1 = disease (original values 1-4 all become 1)
df = raw_df.copy()
df = df.replace('?', np.nan)

df['num'] = pd.to_numeric(df['num'], errors='coerce')
df['num'] = df['num'].apply(lambda x: 0 if x == 0 else 1)

# Drop ID column - it is just a row number, not a medical feature
if 'id' in df.columns:
    df = df.drop(columns=['id'])

print("Target distribution:")
print(df['num'].value_counts())

# ── SPLIT X AND y 
# X = all features going INTO the model
# y = the answer we want the model to predict (heart disease yes or no)
X = df.drop(columns=['num'])
y = df['num']

print(f"\nX shape: {X.shape}")
print(f"y shape: {y.shape}")

# ── BUILD PREPROCESSING PIPELINE ───────────────────────────────────────────



numeric_cols     = X.select_dtypes(include=['int64','float64']).columns.tolist()
categorical_cols = X.select_dtypes(include=['object','category','bool']).columns.tolist()

print(f"\nNumeric columns:     {numeric_cols}")
print(f"Categorical columns: {categorical_cols}")

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler',  StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot',  OneHotEncoder(handle_unknown='ignore'))
])

preprocess = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_cols),
    ('cat', categorical_transformer, categorical_cols)
], remainder='drop')

print("\nPreprocessing pipeline ready")
print("X, y, preprocess all set up correctly")
```

```python
# This function takes any model, runs it through cross-validation,
# and records all the results we need for comparison.
# I used it for every model so the comparison is always fair.
# Ai assisted code 

def evaluate_model(name, model, X, y, cv):
    
    fold_accuracies, fold_precisions = [], []
    fold_recalls, fold_f1s = [], []
    fold_aucs, fold_train_times = [], []
    fold_test_times = []

    for train_idx, test_idx in cv.split(X, y):
        
        # Split into training and test for this fold
        X_train_fold = X.iloc[train_idx]
        X_test_fold  = X.iloc[test_idx]
        y_train_fold = y.iloc[train_idx]
        y_test_fold  = y.iloc[test_idx]

        # Build a fresh pipeline for this fold
        pipe = Pipeline(steps=[
            ('preprocess', preprocess),
            ('model', clone(model))
        ])

        # Train and time it
        start = time.time()
        pipe.fit(X_train_fold, y_train_fold)
        train_time = time.time() - start

        # Test and time it
        start = time.time()
        y_pred = pipe.predict(X_test_fold)
        test_time = time.time() - start

        # Get probabilities for ROC-AUC
        y_prob = pipe.predict_proba(X_test_fold)[:, 1]

        # Record all metrics for this fold
        fold_accuracies.append(accuracy_score(y_test_fold, y_pred))
        fold_precisions.append(precision_score(y_test_fold, y_pred, zero_division=0))
        fold_recalls.append(recall_score(y_test_fold, y_pred, zero_division=0))
        fold_f1s.append(f1_score(y_test_fold, y_pred, zero_division=0))
        fold_aucs.append(roc_auc_score(y_test_fold, y_prob))
        fold_train_times.append(train_time)
        fold_test_times.append(test_time)

    # Return the average across all 5 folds
    return {
        'Model':            name,
        'Accuracy (mean)':  np.mean(fold_accuracies),
        'Precision (mean)': np.mean(fold_precisions),
        'Recall (mean)':    np.mean(fold_recalls),
        'F1 (mean)':        np.mean(fold_f1s),
        'ROC-AUC (mean)':   np.mean(fold_aucs),
        'Accuracy (std)':   np.std(fold_accuracies),
        'Train Time (s)':   np.mean(fold_train_times),
        'Test Time (s)':    np.mean(fold_test_times)
    }

print("evaluate_model function ready")
```

```python
# Set up 5-fold stratified cross-validation
# Stratified means each fold keeps the same balance of sick/healthy patients
# This is important because our dataset has slightly more disease cases than healthy
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

# These are our 4 single models - no ensembles here


single_models = {
    'LogReg':       LogisticRegression(max_iter=2000, random_state=RANDOM_STATE),
    'DecisionTree': DecisionTreeClassifier(random_state=RANDOM_STATE),
    'SVM-RBF':      SVC(kernel='rbf', probability=True, random_state=RANDOM_STATE),
    'NaiveBayes':   GaussianNB()
}

# Run each model through the evaluation function and collect results
print("Running single model comparison...")
print()

baseline_rows = []

for name, model in single_models.items():
    print(f"  Testing {name}...")
    result = evaluate_model(name, model, X, y, cv)
    baseline_rows.append(result)

# Build the results table and sort by ROC-AUC
baseline_results = pd.DataFrame(baseline_rows)
baseline_results = baseline_results.sort_values(
    'ROC-AUC (mean)', ascending=False
).reset_index(drop=True)

print()
print("SINGLE MODEL RESULTS:")
display(baseline_results.round(4))

# Save the best single model name for later comparison
best_single_name = baseline_results.loc[0, 'Model']
print(f"\nBest single model: {best_single_name}")
```

```python
# HYPERPARAMETER TUNING
# for SVM and Decision Tree before the final comparison.
# Running models on default settings is not a fair test.
# GridSearchCV tries every combination and picks the best one.

from sklearn.model_selection import GridSearchCV

print("Tuning SVM...")


svm_pipeline = Pipeline(steps=[
    ('preprocess', preprocess),
    ('model', SVC(kernel='rbf', probability=True, random_state=RANDOM_STATE))
])

svm_param_grid = {
    'model__C':     [0.1, 1, 10, 100],
    'model__gamma': ['scale', 'auto', 0.01, 0.001]
}

svm_search = GridSearchCV(
    svm_pipeline,
    svm_param_grid,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE),
    scoring='roc_auc',
    n_jobs=-1
)

svm_search.fit(X, y)

print(f"  Best SVM settings: {svm_search.best_params_}")
print(f"  Best SVM ROC-AUC:  {svm_search.best_score_:.4f}")

# ── DECISION TREE TUNING ─────────
print("\nTuning Decision Tree...")

# Decision Tree tuning - tried different tree depths and split sizes
# max_depth controls how many questions the tree can ask
# min_samples_split controls how many patients needed before splitting
dt_pipeline = Pipeline(steps=[
    ('preprocess', preprocess),
    ('model', DecisionTreeClassifier(random_state=RANDOM_STATE))
])

dt_param_grid = {
    'model__max_depth':        [3, 5, 7, 10, None],
    'model__min_samples_split': [2, 5, 10]
}

dt_search = GridSearchCV(
    dt_pipeline,
    dt_param_grid,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE),
    scoring='roc_auc',
    n_jobs=-1
)

dt_search.fit(X, y)

print(f"  Best Decision Tree settings: {dt_search.best_params_}")
print(f"  Best Decision Tree ROC-AUC:  {dt_search.best_score_:.4f}")

# ── SAVE BEST MODELS ─────

best_svm_params = {
    k.replace('model__', ''): v
    for k, v in svm_search.best_params_.items()
}

best_dt_params = {
    k.replace('model__', ''): v
    for k, v in dt_search.best_params_.items()
}

best_svm = SVC(
    kernel='rbf',
    probability=True,
    random_state=RANDOM_STATE,
    **best_svm_params
)

best_dt = DecisionTreeClassifier(
    random_state=RANDOM_STATE,
    **best_dt_params
)

print("\nTuning complete")
print(f"Best SVM:           {best_svm_params}")
print(f"Best Decision Tree: {best_dt_params}")
```

```python
# RE-RUN BASELINE WITH TUNED MODELS
# I re-ran the comparison so everything is fairly tuned before
# we compare against ensembles.

print("Re-running baseline with tuned models...")
print()

tuned_models = {
    'LogReg': LogisticRegression(
        max_iter=2000,
        random_state=RANDOM_STATE
    ),
    'DecisionTree (tuned)': DecisionTreeClassifier(
        max_depth=3,
        min_samples_split=2,
        random_state=RANDOM_STATE
    ),
    'SVM-RBF (tuned)': SVC(
        kernel='rbf',
        C=10,
        gamma=0.01,
        probability=True,
        random_state=RANDOM_STATE
    ),
    'NaiveBayes': GaussianNB()
}

tuned_baseline_rows = []

for name, model in tuned_models.items():
    print(f"  Testing {name}...")
    result = evaluate_model(name, model, X, y, cv)
    tuned_baseline_rows.append(result)

# Build the tuned baseline results table
tuned_baseline = pd.DataFrame(tuned_baseline_rows)
tuned_baseline = tuned_baseline.sort_values(
    'ROC-AUC (mean)', ascending=False
).reset_index(drop=True)

print()
print("TUNED BASELINE RESULTS:")
display(tuned_baseline.round(4))

best_single_name = tuned_baseline.loc[0, 'Model']
print(f"\nBest single model after tuning: {best_single_name}")
```

```python


# THE 6 ENSEMBLE EXPERIMENTS
# I test 3 generative strategies x 2 consensus strategies = 6 experiments


def evaluate_model_hard(name, model, X, y, cv):
    fold_accuracies, fold_precisions = [], []
    fold_recalls, fold_f1s = [], []
    fold_train_times, fold_test_times = [], []

    for train_idx, test_idx in cv.split(X, y):
        X_train_fold = X.iloc[train_idx]
        X_test_fold  = X.iloc[test_idx]
        y_train_fold = y.iloc[train_idx]
        y_test_fold  = y.iloc[test_idx]

        pipe = Pipeline(steps=[
            ('preprocess', preprocess),
            ('model', clone(model))
        ])

        start = time.time()
        pipe.fit(X_train_fold, y_train_fold)
        train_time = time.time() - start

        start = time.time()
        y_pred = pipe.predict(X_test_fold)
        test_time = time.time() - start

        fold_accuracies.append(accuracy_score(y_test_fold, y_pred))
        fold_precisions.append(precision_score(y_test_fold, y_pred, zero_division=0))
        fold_recalls.append(recall_score(y_test_fold, y_pred, zero_division=0))
        fold_f1s.append(f1_score(y_test_fold, y_pred, zero_division=0))
        fold_train_times.append(train_time)
        fold_test_times.append(test_time)

    return {
        'Model':            name,
        'Accuracy (mean)':  np.mean(fold_accuracies),
        'Precision (mean)': np.mean(fold_precisions),
        'Recall (mean)':    np.mean(fold_recalls),
        'F1 (mean)':        np.mean(fold_f1s),
        'ROC-AUC (mean)':   np.nan,
        'Accuracy (std)':   np.std(fold_accuracies),
        'Train Time (s)':   np.mean(fold_train_times),
        'Test Time (s)':    np.mean(fold_test_times)
    }

print("Hard voting evaluator ready")
print()
print("=" * 60)
print("RUNNING 6 ENSEMBLE EXPERIMENTS")
print("=" * 60)

ensemble_rows = []

# ── EXPERIMENT 1 
# Generative: Different Algorithms
# Consensus:  Soft Voting
print("\nExperiment 1: Different Algorithms + Soft Voting...")

exp1_model = VotingClassifier(
    estimators=[
        ('lr',  LogisticRegression(max_iter=2000, random_state=RANDOM_STATE)),
        ('dt',  DecisionTreeClassifier(max_depth=3, random_state=RANDOM_STATE)),
        ('svm', SVC(kernel='rbf', C=10, gamma=0.01,
                    probability=True, random_state=RANDOM_STATE)),
        ('nb',  GaussianNB())
    ],
    voting='soft'
)

result = evaluate_model(
    'Exp1: Diff Algorithms + Soft',
    exp1_model, X, y, cv
)
ensemble_rows.append(result)
print(f"  ROC-AUC: {result['ROC-AUC (mean)']:.4f}  "
      f"Accuracy: {result['Accuracy (mean)']:.4f}")

# ── EXPERIMENT 2 
# Generative: Different Algorithms
# Consensus:  Hard Voting
# Hard voting = simple majority - no probability scores available
print("\nExperiment 2: Different Algorithms + Hard Voting...")

exp2_model = VotingClassifier(
    estimators=[
        ('lr',  LogisticRegression(max_iter=2000, random_state=RANDOM_STATE)),
        ('dt',  DecisionTreeClassifier(max_depth=3, random_state=RANDOM_STATE)),
        ('svm', SVC(kernel='rbf', C=10, gamma=0.01,
                    probability=True, random_state=RANDOM_STATE)),
        ('nb',  GaussianNB())
    ],
    voting='hard'
)

result = evaluate_model_hard(
    'Exp2: Diff Algorithms + Hard',
    exp2_model, X, y, cv
)
ensemble_rows.append(result)
print(f"  Accuracy: {result['Accuracy (mean)']:.4f}  "
      f"(ROC-AUC not available for hard voting)")

# ── EXPERIMENT 3 
# Generative: Bootstrap Sampling
# Consensus:  Soft Voting
# Classical bagging - same algorithm, different random data samples
print("\nExperiment 3: Bootstrap Sampling + Soft Voting...")

exp3_model = BaggingClassifier(
    estimator=DecisionTreeClassifier(max_depth=3, random_state=RANDOM_STATE),
    n_estimators=50,
    bootstrap=True,
    random_state=RANDOM_STATE,
    n_jobs=-1
)

result = evaluate_model(
    'Exp3: Bootstrap + Soft',
    exp3_model, X, y, cv
)
ensemble_rows.append(result)
print(f"  ROC-AUC: {result['ROC-AUC (mean)']:.4f}  "
      f"Accuracy: {result['Accuracy (mean)']:.4f}")

# ── EXPERIMENT 4 
# Generative: Bootstrap Sampling
# Consensus:  Hard Voting
print("\nExperiment 4: Bootstrap Sampling + Hard Voting...")

exp4_model = BaggingClassifier(
    estimator=DecisionTreeClassifier(max_depth=3, random_state=RANDOM_STATE),
    n_estimators=50,
    bootstrap=True,
    random_state=RANDOM_STATE,
    n_jobs=-1
)

result = evaluate_model_hard(
    'Exp4: Bootstrap + Hard',
    exp4_model, X, y, cv
)
ensemble_rows.append(result)
print(f"  Accuracy: {result['Accuracy (mean)']:.4f}  "
      f"(ROC-AUC not available for hard voting)")

# ── EXPERIMENT 5 
# Generative: Different Parameters (same algorithm, different depths)
# Consensus:  Soft Voting
print("\nExperiment 5: Different Parameters + Soft Voting...")

exp5_model = VotingClassifier(
    estimators=[
        ('dt_shallow', DecisionTreeClassifier(
            max_depth=2, random_state=RANDOM_STATE)),
        ('dt_medium',  DecisionTreeClassifier(
            max_depth=5, random_state=RANDOM_STATE)),
        ('dt_deep',    DecisionTreeClassifier(
            max_depth=10, random_state=RANDOM_STATE)),
        ('dt_full',    DecisionTreeClassifier(
            max_depth=None, random_state=RANDOM_STATE))
    ],
    voting='soft'
)

result = evaluate_model(
    'Exp5: Diff Parameters + Soft',
    exp5_model, X, y, cv
)
ensemble_rows.append(result)
print(f"  ROC-AUC: {result['ROC-AUC (mean)']:.4f}  "
      f"Accuracy: {result['Accuracy (mean)']:.4f}")

# ── EXPERIMENT 6 
# Generative: Different Parameters (same trees as Exp 5)
# Consensus:  Hard Voting
print("\nExperiment 6: Different Parameters + Hard Voting...")

exp6_model = VotingClassifier(
    estimators=[
        ('dt_shallow', DecisionTreeClassifier(
            max_depth=2, random_state=RANDOM_STATE)),
        ('dt_medium',  DecisionTreeClassifier(
            max_depth=5, random_state=RANDOM_STATE)),
        ('dt_deep',    DecisionTreeClassifier(
            max_depth=10, random_state=RANDOM_STATE)),
        ('dt_full',    DecisionTreeClassifier(
            max_depth=None, random_state=RANDOM_STATE))
    ],
    voting='hard'
)

result = evaluate_model_hard(
    'Exp6: Diff Parameters + Hard',
    exp6_model, X, y, cv
)
ensemble_rows.append(result)
print(f"  Accuracy: {result['Accuracy (mean)']:.4f}  "
      f"(ROC-AUC not available for hard voting)")

# ── FINAL ENSEMBLE RESULTS TABLE 
ensemble_results = pd.DataFrame(ensemble_rows)

print()
print("=" * 60)
print("ALL 6 EXPERIMENTS COMPLETE")
print("=" * 60)
display(ensemble_results.round(4))
```

```python

# Rebuild final_comparison first
final_comparison = pd.concat(
    [tuned_baseline, ensemble_results],
    ignore_index=True
)

final_comparison = final_comparison.sort_values(
    'ROC-AUC (mean)',
    ascending=False,
    na_position='last'
).reset_index(drop=True)

# ── CHART 1: ACCURACY COMPARISON ──────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

acc_plot = final_comparison.sort_values('Accuracy (mean)', ascending=True)
is_ensemble_acc = acc_plot['Model'].str.startswith('Exp')
colours_acc = ['#e74c3c' if e else '#3498db' for e in is_ensemble_acc]

axes[0].barh(
    acc_plot['Model'],
    acc_plot['Accuracy (mean)'],
    color=colours_acc
)
axes[0].set_title('Accuracy Comparison\n(Blue = Single Model, Red = Ensemble)',
                  fontsize=12)
axes[0].set_xlabel('Mean Accuracy')
axes[0].axvline(x=0.80, color='gray', linestyle='--', alpha=0.5)

roc_plot = final_comparison.dropna(subset=['ROC-AUC (mean)']).sort_values(
    'ROC-AUC (mean)', ascending=True
)
roc_colours = ['#e74c3c' if e else '#3498db'
               for e in roc_plot['Model'].str.startswith('Exp')]

axes[1].barh(
    roc_plot['Model'],
    roc_plot['ROC-AUC (mean)'],
    color=roc_colours
)
axes[1].set_title('ROC-AUC Comparison\n(Blue = Single Model, Red = Ensemble)',
                  fontsize=12)
axes[1].set_xlabel('Mean ROC-AUC')
axes[1].axvline(x=0.85, color='gray', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig('final_comparison_charts.png', dpi=150, bbox_inches='tight')
plt.show()
print("Charts saved as final_comparison_charts.png")

# ── CHART 2: SOFT vs HARD VOTING COMPARISON ───────────────────

soft_experiments = [
    'Exp1: Diff Algorithms + Soft',
    'Exp3: Bootstrap + Soft',
    'Exp5: Diff Parameters + Soft'
]
hard_experiments = [
    'Exp2: Diff Algorithms + Hard',
    'Exp4: Bootstrap + Hard',
    'Exp6: Diff Parameters + Hard'
]

soft_acc = final_comparison[
    final_comparison['Model'].isin(soft_experiments)
]['Accuracy (mean)'].values

hard_acc = final_comparison[
    final_comparison['Model'].isin(hard_experiments)
]['Accuracy (mean)'].values

labels = ['Diff Algorithms', 'Bootstrap', 'Diff Parameters']
x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 5))
bars1 = ax.bar(x - width/2, soft_acc, width,
               label='Soft Voting', color='#2ecc71')
bars2 = ax.bar(x + width/2, hard_acc, width,
               label='Hard Voting', color='#e67e22')

ax.set_title('Soft Voting vs Hard Voting\nAcross All Generative Strategies',
             fontsize=12)
ax.set_ylabel('Mean Accuracy')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
ax.set_ylim(0.70, 0.90)

for bar in bars1:
    ax.text(bar.get_x() + bar.get_width()/2,
            bar.get_height() + 0.003,
            f'{bar.get_height():.3f}',
            ha='center', va='bottom', fontsize=9)

for bar in bars2:
    ax.text(bar.get_x() + bar.get_width()/2,
            bar.get_height() + 0.003,
            f'{bar.get_height():.3f}',
            ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.savefig('soft_vs_hard_voting.png', dpi=150, bbox_inches='tight')
plt.show()
print("Soft vs Hard chart saved as soft_vs_hard_voting.png")
```

```python
# CONFUSION MATRICES
# They show exactly how the model performed on each class:

# showing confusion matrices for:
# 1. All 4 tuned single models
# 2. The best ensemble (Exp1: Different Algorithms + Soft Voting)

print("Generating confusion matrices...")
print()

# We need a single train/test split for confusion matrices
# Using the same random state so results are consistent
X_train_cm, X_test_cm, y_train_cm, y_test_cm = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE
)

# Define the models we want confusion matrices for
cm_models = {
    'LogReg': LogisticRegression(
        max_iter=2000, random_state=RANDOM_STATE),
    'DecisionTree (tuned)': DecisionTreeClassifier(
        max_depth=3, min_samples_split=2, random_state=RANDOM_STATE),
    'SVM-RBF (tuned)': SVC(
        kernel='rbf', C=10, gamma=0.01,
        probability=True, random_state=RANDOM_STATE),
    'NaiveBayes': GaussianNB(),
    'Best Ensemble\n(Diff Algorithms + Soft)': VotingClassifier(
        estimators=[
            ('lr',  LogisticRegression(
                max_iter=2000, random_state=RANDOM_STATE)),
            ('dt',  DecisionTreeClassifier(
                max_depth=3, random_state=RANDOM_STATE)),
            ('svm', SVC(kernel='rbf', C=10, gamma=0.01,
                        probability=True, random_state=RANDOM_STATE)),
            ('nb',  GaussianNB())
        ],
        voting='soft'
    )
}

# Plot all confusion matrices in a grid
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.flatten()

for idx, (name, model) in enumerate(cm_models.items()):

    # Build pipeline with preprocessing
    pipe = Pipeline(steps=[
        ('preprocess', preprocess),
        ('model', model)
    ])

    # Train and predict
    pipe.fit(X_train_cm, y_train_cm)
    y_pred_cm = pipe.predict(X_test_cm)

    # Calculate confusion matrix values
    cm = confusion_matrix(y_test_cm, y_pred_cm)
    tn, fp, fn, tp = cm.ravel()

    # Plot
    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm,
        display_labels=['No Disease', 'Disease']
    )
    disp.plot(ax=axes[idx], colorbar=False, cmap='Blues')
    axes[idx].set_title(
        f'{name}\n'
        f'Accuracy: {(tp+tn)/(tp+tn+fp+fn):.3f} | '
        f'Recall: {tp/(tp+fn):.3f}',
        fontsize=10
    )

    # Print the key numbers
    print(f"{name.replace(chr(10), ' ')}")
    print(f"  True Positives  (correctly caught disease):    {tp}")
    print(f"  True Negatives  (correctly cleared healthy):   {tn}")
    print(f"  False Positives (wrongly flagged as disease):  {fp}")
    print(f"  False Negatives (missed disease - DANGEROUS):  {fn}")
    print()

# Hide the last empty subplot
axes[5].set_visible(False)

plt.suptitle(
    'Confusion Matrices: Single Models vs Best Ensemble\n'
    '(False Negatives are most critical in cardiac screening)',
    fontsize=13, y=1.02
)
plt.tight_layout()
plt.savefig('confusion_matrices.png', dpi=150, bbox_inches='tight')
plt.show()
print("Confusion matrices saved as confusion_matrices.png")
```
